```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="ticks")
sns.set_style("darkgrid")
```

```python
def plot_loss(log_file):
    losses = []
    with open(log_file, "r") as _f:
        for line in _f.readlines():
            loss = float(line.split(" ")[3])
            losses.append(loss)

    # print(losses)
    plt.plot(np.arange(len(losses)), losses, label=log_file.rstrip(".log"), alpha=0.7)
    plt.legend()

def plot_flops(log_file):
    flops_list = []
    with open(log_file, "r") as _f:
        for line in _f.readlines():
            flops = float(line.split(" ")[5])
            flops_list.append(flops)

    # print(losses)
    x = np.arange(len(flops_list))
    plt.plot(x, np.ones(len(flops_list)) * np.mean(flops_list))
    plt.plot(x, flops_list, label=log_file.rstrip(".log"), alpha=0.7)
    plt.legend()
```

```python
plot_loss("nn.c.log")
plot_loss("pytorch.log")
plt.savefig("figs/nn.c vs pytorch loss (CPU).pdf")
```

```python
plot_flops("nn.c.log")
plot_flops("pytorch.log")
plt.savefig("figs/nn.c vs pytorch flops (CPU).pdf")
```
