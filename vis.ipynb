```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_loss(log_file):
    losses = []
    with open(log_file, "r") as _f:
        for line in _f.readlines():
            loss = float(line.split(" ")[-1])
            losses.append(loss)

    # print(losses)
    plt.plot(np.arange(len(losses)), losses)
    # plt.show()
```

```python
plot_loss("nn.c.log")
plot_loss("pytorch.log")
```
